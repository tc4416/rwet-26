Based on the examples, the structure of this kind of posts are commonly: \
**1. Opening sentence:** I am ____ (happy/excited/thrilled) to announce that I _______________(did something) \
**2. Thank peple:** Thanks to my ____(mentor/family/professor)\
**3. Reflection:** During my time at ____(previous job/school/bootcamp), I ...\
**4. Excitement / look forward to:** I am excited to start ____  \
**5. Praise the company:** __'s mission to ____?? \
**Hashtag:** #company name, #job Field, #newGrad, #intersnhip #PhD etc\

with point 2-5 arrange different



I think this is a good application on tracery, so I decided to start from making tracery. I think I might need to have sentence structure tracery inside paragraph structure tracery and I am not sure if it will work, so I start by trying it out with just simple open sentence.

Also I want to explore the possibility of having random order of point 2-5 but I will start from one type of 
order for now. 

```python
import tracery
from tracery.modifiers import base_english
import random
rules = {
    "origin": "#open_sentence#",
    "open_sentence": "I am #emotion# to #achievement# at #company# as #position.a#" ,
    "emotion" : ["thrilled", "excited", "happy", "honored", "psyched"],
    "achievement" : ["share that I will be joining a new team", "announce that I’m starting a new role","share that I have accepted the job offer","take on this exciting opportunity"],
    "company": ["Google", "Microsoft", "OpenAI", "a startup"],
    "position": ["Software Engineer", "Data Scientist", "Sushi Cook", "Industrial Design Intern", "Associate Product Manager", "HR Specialist", "IT Manager", "Operations Manager,"]

}

grammar = tracery.Grammar(rules)
grammar.add_modifiers(base_english)
print(grammar.flatten("#origin#"))

```
