# Background
I was thinking about what theme I want to use for this assignment when I walk to school (I live just a few block a way on Albee Sqr in downtown area ), and I noticed that my thoughts are jumping from one topic to another very quickly, along with interuption caused by everything happening on the street, the stores I pass by, the loud noise etc.. So I think maybe I could create a generative poem they represent this chaotic/anxious 7 minutes walk I have everyday on my way to school.

This code is fairly simple and is based on "A House of Dust" poetry generator by Alison Knowles and James Tenney. "A House of Dust" seperate the swappable words/sentences into 4 categaries : materials, locations, light, "inhabitant", and I am using the same technique here. What is different is that I put phrase that will use multiple times into funtions so that they're easier to be reused. I chose the words purely based on whatever comes into my mind.

```python
store = ["pizza shop", "coffee shop", "Pio bagels", "secondhand cigarette", "dispensary" , "construction", "liquor store", "sweetgreen", "blue sky", "boba shop", "chipotle"]
```

```python
problems = ["need to find a job", "am behind on everything","am late", "feel good today", "have five things due today", "don't even want to be here", "am so tired", "want to stay home and sleep", "feel grateful for all I have", "am just existing", "am just walking"]
```

```python
adj = ["chaotic",  "human", "beautiful", "poetic", "romantic", "precious", "sweet"]
```

```python
verb = ["have so much\nto do", "be alive", "be here", "care so much", "be a human", "feel so deeply", "have struggles"]
```

```python
theyWill = ["ask me what I think", "tell me what to do", "want me to speak up", "get me to talk about myself", "want to know what is on my mind"]
```

```python
#text operation
import random

output = ""

def block():
    block_store = ""
    for i in range(4):
         block_store += random.choice(store) + ", ";
    return block_store;

def they():
    theySay= "";
    index = random.sample(range(0,len(theyWill)),3)
    for i in range(3):
        theySay+= "they will " + theyWill[index[i]] + "\n"

    return theySay
                    

def thought1():
    thought= "and I " + random.choice(problems) + "\n\n"
    return thought

 
thought2 = they()

thought3 = "how " + random.choice(adj) +" it is to \n" + random.choice(verb);

output = block() + "\n" + thought1()+ block() +"\nfrom Albee Square to Jay Street\n"+ thought1() + "but \n"+ thought2 + "SOFTLY! \n \n" + block() +"\n"+ thought3;
print(output)
```

## A House of Dust

Original written in Fortran in 1967 by Alison Knowles and James Tenney. [ELMCIP entry](https://elmcip.net/creative-work/house-dust). [More information](http://blog.calarts.edu/2009/09/10/alison-knowles-james-tenney-and-the-house-of-dust-at-calarts/). [Watch Alison Knowles read from this piece](https://www.youtube.com/watch?v=-68Z708lFsY).

```python
import random
```

```python
materials = [
    'brick',
    'broken dishes',
    'discarded clothing',
    'dust',
    'glass',
    'leaves',
    'mud',
    'paper',
    'plastic',
    'roots',
    'sand',
    'steel',
    'stone',
    'straw',
    'tin',
    'weeds',
    'wood'
]
```

```python
locations = [
    'among high mountains',
    'among other houses',
    'among small hills',
    'by a river',
    'by an abandoned lake',
    'by the sea',
    'in a cold, windy climate',
    'in a deserted airport',
    'in a deserted church',
    'in a deserted factory',
    'in a green, mossy terrain',
    'in a hot climate',
    'in a metropolis',
    'in a place with both heavy rain and bright sun',
    'in an overpopulated area',
    'in dense woods',
    'in heavy jungle undergrowth',
    'in japan',
    'in michigan',
    'in southern france',
    'inside a mountain',
    'on an island',
    'on the sea',
    'underwater'
]
```

```python
lights = [
    'all available lighting',
    'candles',
    'electricity',
    'natural light'
]
```

```python
inhabitants = [
    'all races of men represented wearing predominantly red clothing',
    'children and old people',
    'collectors of all types',
    'fishermen and families',
    'french and german speaking people',
    'friends',
    'friends and enemies',
    'horses and birds',
    'little boys',
    'lovers',
    'people from many walks of life',
    'people speaking many languages wearing little or no clothing',
    'people who eat a great deal',
    'people who enjoy eating together',
    'people who love to read',
    'people who sleep almost all the time',
    'people who sleep very little',
    'various birds and fish',
    'vegetarians',
    'very tall people'
]
```

```python
stanza_count = 7
for i in range(stanza_count):
    print()
    print("A house of " + random.choice(materials))
    print("     " + random.choice(locations))
    print("          using " + random.choice(lights))
    print("                inhabited by " + random.choice(inhabitants))
```
